```python
from ultralytics import YOLO

model = YOLO("yolo11m.pt")
model.conf = 0.4
model.iou = 0.45
model.max_det = 100

results = model("/content/drive/MyDrive/computervision/13476456_3840_2160_120fps.mp4", save=True, imgsz=640, vid_stride=12)

```

```python
results = model("/content/runs/detect/predict4/13476456_3840_2160_120fps.avi", save =True)
```

```python
#bot sort tracker >>> by deafult it will take bot sort tracker
results = model.track(source ="/content/drive/MyDrive/computervision/13476456_3840_2160_120fps.mp4",persist = True,save = True)
```
